```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.info() #fir each column gives the info, total entries, total rows, total column, etc
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0") # beacuse we have blank spaces in this column this is why it was showing object
df["TotalCharges"]= df["TotalCharges"].astype("float") # converted data type to float cause max values we in decimal 
```

```python
# we are converting values for senior citizen in the form of yes no for our convience
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"]= df["SeniorCitizen"].apply(conv) # only to make it easier to understand
```

```python
df.head()
```

```python
# we need the count also for this data
ax=sns.countplot(x='Churn', data= df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by churn")
plt.show() # shows how many of them have churned and how many have not
```

```python
# for checking values in percentage
plt.figure(figsize=(4,4))
gb= df.groupby('Churn').agg({'Churn': "count"})
gb
# we will get a yes or no value, for getting percentage after index is added)
plt.pie(gb['Churn'], labels = gb.index, autopct="%1.2f%%") 
plt.title("Percentage of Churned customers")
plt.show()
```

#from the given piechart we can conclude that 26.54% customers have Churned the data
# now lets explore the reason behind it

```python
# first check on the basis of gender
plt.figure(figsize=(4,4))
sns.countplot(x="gender", data=df, hue= "Churn")
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x="SeniorCitizen", data=df)
ax.bar_label(ax.containers[0])
plt.title("no. of senior citizen")
plt.show()
```

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)   #bar chart on the basis of senior citizen
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 2: Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 4))

bottom = [0] * len(percentages)

for churn_value in percentages.columns:
    ax.bar(
        percentages.index,
        percentages[churn_value],
        bottom=bottom,
        label=f"Churn: {churn_value}"
    )
    # Add labels
    for i, (pct, btm) in enumerate(zip(percentages[churn_value], bottom)):
        if pct > 0:
            ax.text(i, btm + pct / 2, f"{pct:.1f}%", ha='center', va='center', fontsize=8)
    # Update bottom for stacking
    bottom = [btm + pct for btm, pct in zip(bottom, percentages[churn_value])]

# Step 3: Customize plot
ax.set_title("Churn by SENIOR CITIZEN")
ax.set_xlabel("Senior Citizen (0 = No, 1 = Yes)")
ax.set_ylabel("Percentage of Customers")
ax.set_xticks([0, 1])
ax.set_xticklabels(["No", "Yes"])
ax.legend(title="Churn") # bbox se we can set the coordinates of the churn
plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up subplot grid
n_cols = 3
n_rows = (len(cols) + n_cols-1) // n_cols # ceiling division
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
axes = axes.flatten()  # flatten in case it's a 2D array

# Plot each countplot
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'count plot of{col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('count')

# Hide any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
